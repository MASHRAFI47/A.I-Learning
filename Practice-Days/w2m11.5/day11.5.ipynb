### 1 . Create a 1D, 2D, and 3D NumPy array and print their dimensions and shapes.


```python
import numpy as np

arr = np.random.randint(1, 10, size=(5,))
arr2 = np.random.randint(1, 10, size=(5,4))
arr3 = np.random.randint(1, 10, size=(3,3,3))

print(arr.ndim)
print(arr.shape)
print(arr2.ndim)
print(arr2.shape)
print(arr3.ndim)
print(arr3.shape)
```

### 2 . Create:
an array of zeros of shape (3,3)


an array of ones of shape (2,5)


an array filled with 7


```python
import numpy as np

arr = np.zeros((3,3))
arr2 = np.ones((2,5))

print(arr)
print(arr2)

arr3 = np.full((3,3), 7)
print(arr3)
```

### 3 . Create an array of even numbers between 10 and 50 using arange.


```python
import numpy as np

arr = np.arange(10, 50, 2)
print(arr)
```

### 4 . Create a 4×4 matrix from 0–15. Then , extract:

```python
import numpy as np

arr = np.random.randint(0, 16, size=(4,4))

first_row = arr[0,:]
last_col = arr[::,-1]
two_two = arr[1:3, 1:3]

print(first_row)
print(last_col)
print(two_two)
```

### 5 . Add a 1D array [1,2,3] to each row of a 3×3 matrix.


```python
import numpy as np

arr = np.random.randint(1, 10, size=(3,3))

arr2 = np.array([1,2,3])

add_arr = arr+arr2
print(add_arr)
```

### 6 . Given two arrays A and B, compute:
(A + B) / 2


A^2 + B^2



```python
import numpy as np

arr = np.random.randint(1, 10, size=(3,3))
arr2 = np.random.randint(1, 10, size=(3,3))

print((arr+arr2)/2)
print(arr**2 + arr2**2)
```
