### Create a Pandas Series from the list [10, 20, 30, 40, 50]. Print the Series, its index, and data type. Then create a DataFrame with columns "Name", "Age", and "City" containing at least 4 rows of data.

```python
import pandas as pd

df = pd.Series([10, 20, 30, 40, 50])
print(df)
print(df.index)
print(df.dtype)

data = {
    'Name': ['John', 'Alice', 'Bob', 'David'],
    'Age': [25, 30, 22, 28],
    'City': ['Dhaka', 'Chittagong', 'Sylhet', 'Khulna']
}

df2 = pd.DataFrame(data, columns=['Name', 'Age', 'City'], index=[1, 2, 3, 4])
df2
```

### Using any DataFrame, write code to print the number of rows and columns, column data types (.dtypes), and summary statistics (.describe()).

```python
print(df2.columns.size)
print(df2.index.size)
print(df2.describe())
```

### Create a DataFrame using the dictionary below:


#### data = {
    'Product': ['Pen', 'Book', 'Eraser', 'Pencil'],
    'Price': [10, 50, 5, 7],
    'Quantity': [100, 60, 200, 150]
#### }

Print .columns, .index. Access a single column and one full row.


```python
data = {
    'Product': ['Pen', 'Book', 'Eraser', 'Pencil'],
    'Price': [10, 50, 5, 7],
    'Quantity': [100, 60, 200, 150]
}


dict_df = pd.DataFrame(data)
print(dict_df)
print(dict_df.columns)
print(dict_df.index)

print(dict_df['Product'])

print()

print(dict_df.loc[0])
```

```python
dict_df
```

```python
print(dict_df.loc[0])
print(dict_df.iloc[0])
print()

print(dict_df.loc[:, ['Price']])
print(dict_df.iloc[:, 1])

print()
print(dict_df.loc[1:2, ['Product', 'Price']])
print(dict_df.iloc[1:3, 0:2])
```

### Create a 4×3 DataFrame of student names and marks. Drop one column using .drop(), drop one row by index

```python
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David'],
    'Math': [85, 90, 78, 88],
    'Science': [92, 85, 80, 89]
}

df = pd.DataFrame(data)
df.drop(1, inplace=True)
df.drop('Science', axis=1)
```

### Create a DataFrame of employees with columns "Name", "Department", and "Salary". Sort it by "Salary" in ascending order and by "Name" in descending order.


```python
data = {
    'Name': ['John', 'Sarah', 'Michael', 'Emma'],
    'Department': ['HR', 'IT', 'Finance', 'Marketing'],
    'Salary': [40000, 55000, 60000, 45000]
}

df = pd.DataFrame(data)
copy_df = df.sort_values(['Salary', 'Name'], ascending=[1,0])
copy_df
```
