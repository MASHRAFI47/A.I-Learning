### 1. Add New Column and Save File
You have a DataFrame with column Price.
 Create a new column Discounted_Price with (10% discounted price).
 Save the modified DataFrame as discounted_products.csv.

```python
import numpy as np 
import pandas as pd 
import requests 
from io import StringIO 

res = requests.get("https://raw.githubusercontent.com/phitronio/Python-for-ML/refs/heads/main/Practice%20Day%20(1).csv")
df = pd.read_csv(StringIO(res.text))
df['Discounted_Price'] = df['Price'] - df['Price'] * (10/100)

df.to_csv("discounted_products.csv")
```

```python
df
```

### 2. Check Duplicates and Remove Them
Column: EmployeeID
 Tasks:
Count duplicate entries


Remove duplicates


Display cleaned DataFrame


```python
print(df['EmployeeID'].duplicated().sum())
df.drop_duplicates(subset=['EmployeeID'], inplace=True)

df
```

### 3. Handle Missing Values
Columns: Age, Salary

#### Tasks:
Count missing values


Fill missing Age with mean age


Drop rows where Salary is missing


```python
df
```

```python
ageMissing = len(df) - len(df['Age'].dropna())
print(ageMissing)

salaryMissing = len(df) - len(df['Salary'].dropna())
print(salaryMissing)

df['Age'] = df['Age'].fillna(df['Age'].mean())
df.dropna(subset=['Salary'], inplace=True)

df
```

### 4. Apply Statistical Functions
Columns: Marks1, Marks2, Marks3
### Tasks:
Calculate mean, max for each subject


Compute total marks per student


Find standard deviation of total marks


```python
print(df['Marks1'].mean())
print(df['Marks2'].mean())
print(df['Marks3'].mean())

print(df['Marks1'].max())
print(df['Marks2'].max())
print(df['Marks3'].max())


df['total_marks'] = df['Marks1'] + df['Marks2'] + df['Marks3']

print(df['total_marks'].std())

df
```

### 5. Histogram, Bar, Scatter, Pie
### Create:
#### Histogram of Age
#### Scatter plot (Age vs Salary)
#### Scatter plot ( Price vs Discounted Price) 


```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

plt.xlabel("People with different Ages")
plt.ylabel("Counts")
plt.title("Histogram of Age")

plt.hist(df['Age'], color="green", edgecolor="black", label="Age")
plt.legend()
plt.show()
```

```python
plt.xlabel("People with different Ages")
plt.ylabel("Salary")
plt.title("Scatter of Age and Salary")

plt.scatter(df['Age'], df['Salary'], s=50, marker="X")
plt.show()
```

```python
plt.xlabel("Price and Discounted price")
plt.ylabel("Discounted_Price")
plt.title("Scatter of Price and Discounted_Price")

plt.scatter(df['Price'], df['Discounted_Price'], s=50, marker="X")
plt.show()
```
