### Web Scrapping Demo

```python
import numpy as np 
import pandas as pd 
import requests 
from bs4 import BeautifulSoup

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36"
}

res = requests.get('https://en.wikipedia.org/wiki/List_of_companies_of_Bangladesh', headers=headers)

soup = BeautifulSoup(res.text, 'lxml')
#print(soup.prettify())
table = soup.find('table') 
tr = soup.find_all('tr')

company = []
for t in tr[2:]:
    td = t.find_all('td')
    if len(td) > 0:   # only process rows that have data cells
        company.append(td[0].text.strip())


#print(company)


df = pd.DataFrame(
    {
        'company': company,
    }
)


df
```
