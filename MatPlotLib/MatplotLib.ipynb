```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('enrollment_data.csv')

plt.xlabel("Year") 
plt.ylabel("Count of Enrollment")
plt.title("Student Enrollment over the years")
 
plt.plot(df['Year'], df['Programming'], label="Programming", color="red", linewidth=5, linestyle='solid', marker='o', markersize=10)
plt.plot(df['Year'], df['Digital Marketing'], label="Digital Marketing", linewidth=3, linestyle="dashdot")

plt.legend(loc="best") # by default best
```

### Scatter Plots

```python
# to find correlation between features
# postivie correlation
# negative correlation
# no correlation
# outliers ( big difference than usual )
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_IQdata.csv")

df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_IQdata.csv")

df

plt.xlabel("Study Hours")
plt.ylabel("IQ score")
plt.title("Relation between Study hours and IQ Score")   #-> Positive Correlation

plt.scatter(df["Study_Hour"], df["IQ_Score"], label="IQ", color="red", linewidth=3)
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_IQdata.csv")

df

plt.xlabel("Shoe Size")
plt.ylabel("IQ Score")
plt.title("Relation between Shoe Size and IQ Score")   #-> No Correlation

plt.scatter(df["Shoe_Size"], df["IQ_Score"], label="IQ", color="green", linewidth=3)
plt.legend()
plt.grid()
plt.show()
```

```python
# as shoe size has no correlation so we can drop it
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_IQdata.csv")

plt.xlabel("Chilling time")
plt.ylabel("IQ Score")
plt.title("Relation between Chilling hours and IQ Score")   #-> Negative Correlation

plt.scatter(df["Chilling_Hours"], df["IQ_Score"], label="IQ", color="cyan", s=50, marker="X")
plt.legend()
plt.grid()
plt.show()
```

### Bar Chart

```python
# categorical
```

```python
data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)

df
```

```python
# gender wise

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

gender_group = df.groupby("Gender").size()

gender_group.index
```

```python
gender_group = df.groupby("Gender").size()

gender_group.values
```

```python
plt.xlabel("Genders") 
plt.ylabel("Frequency/ Count")
plt.title("Bar Chart of Gender Among Students")


plt.bar(gender_group.index, gender_group.values, color=["orange", "red"], edgecolor="black")
```

### Pie Chart

```python
# Out of 100

# bar chart on real data

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

df = pd.read_csv("student_data.csv")

completion = df.groupby("CompletionStatus").size()

plt.pie(completion, labels=completion.index, autopct='%1.1f%%', colors=["green", "blue", "red"], explode=(0.1, 0, 0))
plt.show()
```

```python
# Out of 100

# bar chart on real data

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

df = pd.read_csv("student_data.csv")

completion = df.groupby("CompletionStatus").size()

plt.pie(completion, labels=completion.index, autopct='%1.1f%%', colors=["green", "blue", "red"], explode=(0.1, 0, 0), shadow=True)
plt.show()
```
