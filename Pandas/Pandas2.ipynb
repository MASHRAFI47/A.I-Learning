### Filtering based on string

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hannah","Sakib"],
    "City": ["New York", "Los Angeles", "Newark", "Boston", "New Delhi", "Chicago", "New Orleans", "Houston","H Los Ang"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Marketing", "Finance", "HR", "HR"],
    "Salary": [50000, 60000, 55000, 70000, 52000, 58000, 62000, 51000,70000]
}

df = pd.DataFrame(data)
df
```

```python
# get string starting with new

df.loc[df['City'].str.contains("New")]
```

```python
df.loc[df['City'].str.contains("new", case=False)]
```

```python
#starts with Los, not in middle

df.loc[df['City'].str.contains(r"^Los")]
```

```python
#ends with rk

df.loc[df['City'].str.contains(r"rk$")]
```

```python
# name starts with a vowel in list

df.loc[df['Name'].str.contains(r"^[aeiou]", case=False)]
```

```python
df.loc[df['City'].str.contains(r"New|Los")]
```

### Apply Function in Pandas

```python
#min max scaling

mn = df['Total Marks'].min()
mx = df['Total Marks'].max()

print(mn) 
print(mx) 

#scaled = (x - min) / (mx - mn)
```

```python
df['Scaled Marks'] = df['Total Marks'].apply(lambda x : (x-mn) / (mx-mn))

df
```

```python
# custom built function 

def grading_system(marks):
    if marks >= 260:
        return "A+"
    elif marks >= 250:
        return "A"
    else:
        return "A-"

df['Grade'] = df['Total Marks'].apply(grading_system)
```

```python
df
```

```python
def marking_system(df):
    a = df['Data Structure Marks'] * 2
    b = df['Python Marks'] * 3
    c = df['Algorithm Marks'] * 4

    return a+b+c

df['Exceptional Marks'] = df.apply(marking_system, axis=1)

df
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})
                   
#print(df)  

# Drop rows where column A has missing values
print(df.dropna(subset=['A']))


# Drop rows with less than 2 non-null values

print(df.dropna(thresh=2))

# Drop rows where both Column A and Column B has missing values

print(df.dropna(subset=['A', 'B'], axis=0))
```

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
df = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Value1': [10, 20, 30, 40, 50, 60],
    'Value2': [100, 200, 300, 400, 500, 600]
})

# Basic groupby and aggregation
print("Mean of Value1 for each Category:")
print(df.groupby('Category')['Value1'].mean())

print("Max of Value2 for each Category:")
print(df.groupby('Category')['Value2'].max())

print("Sum of Value1 for each Category:")
print(df.groupby('Category')['Value1'].sum())

df

```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})
                   
print(df)  

# Replace null values based on given conditions
df['A'].fillna(0.0, inplace=True)
df['B'].fillna(99.0, inplace=True)
df['C'].fillna("Unknown", inplace=True)
print(df)


df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})
# Forward fill column B

df['B'].fillna(5.0, inplace=True)
print(df)

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})
# Fills missing values in numeric columns with the mean of the column and in string columns with 'Unknown'.
df['A'].fillna(df['A'].mean(),inplace=True)
df['B'].fillna(df['B'].mean(),inplace=True)
df['C'].fillna("Unknown", inplace=True)

print(df)
```
