### Pandas Dataframe from inbuilt DS data

```python
import pandas as dp 

my_list = [['Alice', 25], ['Bob', 30], ['Charlie', 28]]

list_df = pd.DataFrame(my_list, columns=['Name', 'Age'], index=[1,2,3]) 
print(type(list_df))
list_df
```

```python
my_tuple = (('Alice', 25), ('Bob', 30), ('Charlie', 28))

tuple_df = pd.DataFrame(my_tuple, columns=['Name', 'Age'])
print(type(tuple_df))
tuple_df
```

```python
my_dict = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 29,28]
}

dict_df = pd.DataFrame(my_dict)
dict_df
```

```python
my_data = [
    {'Name': 'Alice', 'Age': 25, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 30, 'City': 'Paris'},
    {'Name': 'Charlie', 'Age': 28}
]

my_df = pd.DataFrame(my_data)
my_df
```

### Dropping rows and column, Assigning values and iteration

```python
df
```

```python
#row deletion
df.drop(1, inplace=True)

df
```

```python
#col deletion
df.drop('Instructor', axis=1, inplace=True)

df
```

```python
# Assign values

df.loc[6, 'Python Marks'] = 90
df.loc[6, 'CompletionStatus'] = "Completed"

df.head()
```

```python
# add 2 to first 3 students

df.loc[1:3, 'Python Marks'] += 2
df.head() 
```

```python
# iteration using for loop which is not recommended

for i, series in df.iterrows():
    print(f"{i} : {series}")
```

```python
for i in df.itertuples(index=False):
    print(i)
```

```python
completed = df.loc[df['CompletionStatus']=='Completed']
completed
```
